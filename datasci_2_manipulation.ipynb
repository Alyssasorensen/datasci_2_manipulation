```python
import pandas as pd

data = pd.DataFrame({
    'Bad Row 1': [1, 2],
    'Bad Row 2': [3, 4]
})

print(data)

# Function to remove white space and special characters from a value
def clean_column(column):
    # Convert integers to strings
    column = column.astype(str)

    # Remove white space and special characters, and convert to lowercase
    column = column.str.replace(r'[^a-zA-Z0-9]', '', regex=True)
    column = column.str.lower()
    return column

# Apply the clean_column function to all columns
for column in data.columns:
    data[column] = clean_column(data[column])

print(data)

```
