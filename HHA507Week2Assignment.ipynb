## Load Packages

```python
import pandas as pd
```

## Load Data

```python
healthcare_data_cleaning = pd.read_csv('https://raw.githubusercontent.com/Alyssasorensen/datasci_2_manipulation/main/datasets/raw.githubusercontent.com_hantswilliams_HHA_507_2023_main_WK2_data_healthcare_data_cleaning.csv')
```

```python
healthcare_data_cleaning
```

## Removing Duplicate Rows

```python
healthcare_data_cleaning.drop_duplicates(inplace=True)
healthcare_data_cleaning
```

## Removing Duplicate Columns

```python
healthcare_data_cleaning = healthcare_data_cleaning.T.drop_duplicates().T
```

**## Removing White Space and Special Characters**

```python
import pandas as pd
import re

def clean_value(value):
    cleaned_value = re.sub(r'\s+|[^a-zA-Z0-9]', '', value)
    return cleaned_value

# Apply the clean_value function to all columns
for column in healthcare_data_cleaning.columns:
    healthcare_data_cleaning[column] = healthcare_data_cleaning[column].apply(clean_value)

healthcare_data_cleaning
```

**## Adding the Decimals Back to the "Payment Due" Column and "Prescription Cost" Column**

```python
import pandas as pd

# Sample DataFrame with cleaned money amounts (without decimals)
data = {'Payment Due ($)': [65.45721578126224, 430.68365678679174, 315.0709305262176, 320.2998987723972, 264.2147372473319]}
healthcare_data_cleaning = pd.DataFrame(data)

# Number of decimal places to round
decimal_places = 2

# Function to round numbers to the specified decimal places
def round_to_decimal_places(value, decimal_places):
    return round(value, decimal_places)

# Apply the rounding function to the 'Amount' column
healthcare_data_cleaning['Payment Due ($)'] = healthcare_data_cleaning['Payment Due ($)'].apply(lambda x: round_to_decimal_places(x, decimal_places))

# Display the corrected DataFrame
print(healthcare_data_cleaning)

```

```python
import pandas as pd
import re

# Sample DataFrame with columns containing various data types
data = {'Column1': [65.45721578126224, 'abc 123', 315.0709305262176],
        'Column2': ['def 456', 320.2998987723972, 264.2147372473319]}
healthcare_data_cleaning = pd.DataFrame(data)

def clean_value(value):
    if isinstance(value, (str, bytes)):
        cleaned_value = re.sub(r'\s+|[^a-zA-Z0-9]', '', str(value))
        return cleaned_value
    else:
        return str(value)

# Apply the clean_value function to all columns
for column in healthcare_data_cleaning.columns:
    healthcare_data_cleaning[column] = healthcare_data_cleaning[column].apply(clean_value)

print(healthcare_data_cleaning)
```

**## Handling Missing Values**

```python
healthcare_data_cleaning = healthcare_data_cleaning.replace('missing','Unknown')
healthcare_data_cleaning
```

## Cleaning Column Names

```python
healthcare_data_cleaning.columns = healthcare_data_cleaning.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
healthcare_data_cleaning
```
